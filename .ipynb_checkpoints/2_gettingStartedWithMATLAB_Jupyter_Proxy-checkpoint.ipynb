## MATLAB in a Jupyter Notebook
If you run a MATLAB command for the first time, you will 1) need to select a MATLAB-License that is fit for Online Licensing (or your admins preconfigured the MATLAB-Jupyter-Proxy to talk to a license server) and 2) wait for startup of the MATLAB Computational Engine. Once these steps are done, the "MATLAB Kernel" is running in the background and you can execute MATLAB Code in the notebook:

````matlab
C = membrane;
dim = 31;
count = 0;
for ix = 1:dim,
    for iy = 1:dim,
        count = count + 1;
        X(count) = ix;
        Y(count) = iy;
    end;
end
Z = C(:);
scatter3(X,Y,Z,marker=".")
xlabel("X")
ylabel("Y")
zlabel("Z")
title("L-shaped Membrane")
````

As you can see, we executed MATLAB code. However, the plots are (by design) static images. You _can_ open a "MATLAB in the Browser" window from the Jupyter Lab.
* Open a new Launcher if none is open: File --> New Launcher
* Click on the "Open MATLAB" Icon. A New Browser Tab will open with a "MATLAB Online"-like GUI
* Copy the line ```scatter3(X, Y, Z, marker=".")``` into the command window and execute...
