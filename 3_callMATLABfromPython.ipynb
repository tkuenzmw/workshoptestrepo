## Connect to MATLAB Engine
The MATLAB Engine for Python () allows you to call an existing, working and licensed MATLAB that is installed in the same machine you run your Python-Kernel in.

If the MATLAB Computational Engine is already running (i.e. you opened MATLAB already elsewhere) you can connect to it:

```python
ML = matlab.engine.connect_matlab()
```

You can then use any MATLAB command with calls in the format of "[MATLAB Engine Object].[MATLAB Command]". Conversion of variable types in the inputs are "partially automatic" (for basic types, arrays etc., not for complex variable types).

Let us look at a simple example...

```python
out = ML.gcd(100.0, 80.0, nargout=3) # find smallest common denominator
print(out)
```
